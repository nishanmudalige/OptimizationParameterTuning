```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
df = pd.read_csv("../results/complete_data_and_metadata.csv")
df
```

```python
df = df.rename(columns={
    "expr_length": "expr_leaves_count",
    "expr_depth": "expr_max_depth",
})

feature_cols = ["nvar", 
                "expr_leaves_count", 
                "expr_max_depth", 
                "init_eval_obj_time", 
                "init_eval_grad_time",
                "mem",
                ]
target_col = "time"
df["log_time"] = np.log(df["time"])
df
```

```python
df[feature_cols].describe()
```

```python
df[target_col].describe()
```

```python
np.log(df[target_col]).describe()
```

```python
features = []
features = feature_cols + ["time", "log_time"]
features
```

```python
os.makedirs("../boxplots", exist_ok=True)

for feature in features:
    plt.figure(figsize=(7, 5))              # create a fresh figure
    plt.boxplot(df[feature].dropna())       # plot only this feature
    plt.title(f"Box Plot Showing the Distribution of '{feature}'")
    plt.tight_layout()
    plt.savefig(f"../boxplots/{feature}.pdf", bbox_inches="tight")
    plt.close()                              # close so next loop starts clean
```
