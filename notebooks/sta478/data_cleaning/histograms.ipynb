```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("./datasets/final.csv")
```

### Histogram of log1p(timed_bytes)

```python
x = data["timed_bytes"]
x.describe()

sns.histplot(x=np.log1p(x))
plt.xlabel("log1p(timed_bytes)")
plt.show()
```

```python
x = data["neval_obj"]
x.describe()

sns.histplot(x=np.log1p(x))
plt.xlabel("log1p(neval_obj)")
plt.show()
```

```python
x = data["neval_grad"]
x.describe()

sns.histplot(x=np.log1p(x))
plt.xlabel("log1p(neval_grad)")
plt.show()
```
