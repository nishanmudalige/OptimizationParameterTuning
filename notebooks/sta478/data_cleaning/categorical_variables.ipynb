```python
import pandas as pd
import numpy as np

# min q1(20%), q2(40%), q3(60%), q4(80%), max

# Load your data
df = pd.read_csv('./datasets/final.csv')

# Create quintile categories for stats_elapsed_time
df['time_cat'] = pd.qcut(df['stats_elapsed_time'], 
                          q=5, 
                          labels=['Very Fast', 'Fast', 'Medium', 'Slow', 'Very Slow'])

# Create quintile categories for nvar
df['nvar_cat'] = pd.qcut(df['nvar'], 
                          q=5, 
                          labels=['Very Small', 'Small', 'Medium', 'Large', 'Very Large'])

# Verify the distribution (each should have ~20% of data)
print("\nTime category distribution:")
print(df['time_cat'].value_counts().sort_index())

print("\nNumber of variables category distribution:")
print(df['nvar_cat'].value_counts().sort_index())

# Show the actual cutpoints
print("\nTime quintiles (in seconds):")
print(pd.qcut(df['stats_elapsed_time'], q=5, retbins=True)[1])

print("\nNvar quintiles:")
print(pd.qcut(df['nvar'], q=5, retbins=True)[1])

# Save to new CSV
df.to_csv('./datasets/data_with_categories.csv', index=False)
print("\nSaved to data_with_categories.csv")
```
